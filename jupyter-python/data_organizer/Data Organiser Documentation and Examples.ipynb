## Methods

```
  db.select(fields, where, sort)
    
  db.get_parameter(name, fields, where, sort)
  
  db.get_parameter(where={...}, name='E')
  db.getParameter('E')
  
```

 * *name* required

 * *fields* is optional, if omitted then all fields returned
 * *where* is optional, if omitted then all (approprtiate) records returned
 * *sort* is optional, of ommited then there is not sorting
