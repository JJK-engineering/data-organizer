```python
# test_1a_4 (learning - calculating distance between long/lat points with GeoSPARQL)
# using 'geosparql-test-kolas_1a.ttl' with added point WM2Point
# GeoSPARQL WKT -> coordinates are given as long/lat (NOT lat/long as is more common)
# see also https://stackoverflow.com/questions/18636564/lat-long-or-long-lat
#   WKT standard is: POINT (x y)
#   WSG 84: http://spatialreference.org/ref/epsg/4326/
#   So it is POINT (Lon Lat)

sparql = SPARQLWrapper("http://34.76.178.181:3031/ds/query")
sparql.setMethod('GET')
sparql.setQuery("""
     PREFIX geo:   <http://www.opengis.net/ont/geosparql#> 
     PREFIX ex:    <http://www.example.org/POI#>
     PREFIX sf:    <http://www.opengis.net/ont/sf#>
     PREFIX geof:  <http://www.opengis.net/def/function/geosparql/>
     PREFIX units: <http://www.opengis.net/def/uom/OGC/1.0/>  #referenced in geosparql-jena README

    #SELECT ?geo1 ?geo2 geof:distance(?geo1, ?geo2, units:meter)
    SELECT ?geo1 ?geo2 geof:distance(?geo1, ?geo2, units:kilometer)

    WHERE {
          ex:WMPoint geo:asWKT ?geo1.
          ex:WM2Point geo:asWKT ?geo2.
          } 
""")

sparql.setReturnFormat('csv')
results = sparql.query().convert()

# result is WRONG (or is it a lat/long vs long/lat [WKT] problem?)
#                 (result is approx. correct for 1 deg long at 38.89 N lat)
# result should be approx. 111 km (1 deg latitude)
# can also check using  https://www.movable-type.co.uk/scripts/latlong.html -> 111.2 km
# https://github.com/galbiston/geosparql-jena
#  -> Apache Jena Spatial Functions/WGS84 Geo Predicates
#  seems that jena-geosparql may use lat/long convention
#  "Geo predicates can be converted to Geometry Literals in query 
#  and then used with the GeoSPARQL filter functions." 
#  (a procedure using BIND is given)
```
